* Question-1 write the program even and odd number take input user

```python

number=int(input("Enter a number: "))
if number % 2 == 0:
    print(f"{number} is even.")   
else:
    print(f"{number} is odd.")                                          
```

* Question-3 Write program calculate electricity bill accept number unit from user 


 1. first 100 unit                             . No charge  
 2. next 100 unit                              . 5 rupees per unit
 3. after 200 unit                             .  10 rupees per unit     

```python
unit=int(input("Enter the Electricity Unit: "))
amount=0
if unit < 100:
    amount=0
elif unit > 100 and unit <= 200:
    amount=(unit-100)*5
elif unit > 200:
    amount=500+(unit-200)*10      
else:
    print("Invalid Input") 

print("amount to be paid:", amount)
```
