### OOPs is a programming approach where programs are built using classes and objects to represent real-world entities.

## Main Features of OOPs

Encapsulation – Data aur functions ko ek unit me bandh kar rakhna

Inheritance – Ek class dusri class ke features use kar sakti hai

Polymorphism – Ek method, multiple forms

Abstraction – Important cheezein dikhana, details hide karna

### Explanation of OOPs

OOPs me poora program class aur object par based hota hai:

Class
Class ek blueprint hoti hai, jisme bataya jata hai object ke pass kaun-kaun se data (variables) aur functions (methods) honge.

Object
Object class ka real instance hota hai jo memory me jagah leta hai.

Example (real life):

Class = Car
Objects = BMW, Audi, Swift
