## Accessing Substrings from a String

```python
# Postive String
s='ramakant sharma'

print(s[2])
print(s[3])
print(s[4])
```

```python
# Negtive String
print(s[-1])
print(s[-3])
print(s[-4])
```

```python
# Slicing
a='golusharma'
print(a[0:4])
#0=kha se start karna h
#4=kha tk slice krna or +1 lena h q ki last wala include nhi hota h


```

```python
s = 'hello world'
print(s[2:])
# start se slice krna h to bas itna likh do ki kaha se start krna h or end tk slice kr dega
print(s[:5])
# end tk slice krna h to bas itna likh do ki kaha tk slice kr
```

```python
s = 'hello world'
print(s[6:0:2])
# 3rd value bhi add kr skte h step ke liye jo ki by default 1 hota h
# step ka mtlb kitne kitne karke slice krna h
print(s[6:0:1])
# 6=5 tk jao or 0 tk jao or 1 step karke jao
```

```python
s = 'hello world'
print(s[-1:-6:-1])
print(s[-1:-6:-2])
```
