# file handling

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print("hi my name is ramakant", self.name, "and my age is", self.age) 


p = Person("ramakant", 24)

import pickle
with open("person_data.pkl", "wb") as file:
    pickle.dump(p, file)
```

```python
# pickle load
import pickle
with open("person_data.pkl", "rb") as file:
    p_loaded = pickle.load(file)
p_loaded.display_info()
```
