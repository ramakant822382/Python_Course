# Loop

```python
# while loop
number=int(input("Enter a number: "))
i=1
while i<=10:
    print(i*number)
    i+=1
```

```python
# guess game
import random
number=random.randint(1,100)        
guess=int(input("Enter your guess: "))
counter=0
while guess!=number:
    if guess<number:
        print("Too low!")
    else:
        print("Too high!")
    guess=int(input("Enter your guess: "))
    counter+=1          
print("Congratulations! You guessed the number in", counter, "attempts .")
```
